# Scoring Function

```python
#weights 
ce = [0.1915782478, 0.20880163, 0.1893672373, 0.2180375816, 0.09638690678, 0.2759764396, 0.1825556534]
pf = [0.208360037, 0.2011743018, 0.1967942263, 0.1981649906, 0.136708574, 0.2461256934, 0.2025059741]
pg = [0.2072206793, 0.193809877, 0.1676177688, 0.2171361032, 0.3769396308, 0.1480528393, 0.2148157464]
sf = [0.1813715018, 0.1989051952, 0.2517483496, 0.178773188, 0.155170461, 0.1790472923, 0.1938907056]
sg = [0.2114695341, 0.197308996, 0.194472418, 0.1878881365, 0.2347944275, 0.1507977355, 0.2062319205]

```

```python
score = []
capstone = boxscore.query("matchId == 108956")
capstone.head(25)
```

```python
def score_calc (values : list, weights : list):
    sum = 0
    max = 0
    for i in range(len(values)):
        if values[i] > 1:
            values[i] = 1
        elif values[i] < 0:
            values[i] = 0
        sum = sum + (values[i]*weights[i]*10)
        max = max + (weights[i]*10)
    return (sum/max)*100
```

```python

for index, row in capstone.iterrows():
    if row['name'] == 'Team':
        score.append(1000.0)
    elif row['sec'] == 0:
        score.append(0.0)
    else:
        pn = row['Ps'] / 25
        if (row['FG_A']+row['FT_A']) == 0:
            efg = 0.0
        else:
            efg = ((row['3P_P']*1.5 + row['2P_P'] + row['FT_P']*0.5)/ (row['FG_A']+row['FT_A']))/1.115
        tn = 1 - (row['TOs']/4)
        sbn = (row['STs']+row['BLs'])/2
        an = row['As']/7
        rn = row['Rs']/10
        tim = row['sec'] / 2400
        values = [pn,efg,tn,sbn,an,tim]
        weights = []
        if row['position'] == "C":
            score.append(score_calc(values,ce))
        elif row['position'] == "PF":
            score.append(score_calc(values,pf))
        elif row['position'] == "PG":
            score.append(score_calc(values,pg))
        elif row['position'] == "SF":
            score.append(score_calc(values,sf))
        elif row['position'] == "SG":
            score.append(score_calc(values,sg))
        else:
            score.append(-1)

capstone['scores'] = score
capstone.head(25)            
```
